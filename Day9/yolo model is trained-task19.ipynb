```python
import zipfile

with zipfile.ZipFile("/content/traffic_violation.zip", "r") as zip_ref:
    zip_ref.extractall("/content/dataset")

print("Dataset Extracted Successfully!")
```

```python
import os

for root, dirs, files in os.walk("/content/dataset"):
    if "data.yaml" in files:
        print("FOUND:", os.path.join(root, "data.yaml"))
```

```python
import os

print("Train Images:", len(os.listdir("/content/dataset/train/images")))
print("Valid Images:", len(os.listdir("/content/dataset/valid/images")))
print("Test Images:", len(os.listdir("/content/dataset/test/images")))
```

```python
from ultralytics import YOLO

# Load pretrained YOLOv8 Nano
model = YOLO("yolov8n.pt")

# Train
model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0
)
```

```python
import os

for root, dirs, files in os.walk("/content/runs/detect/train"):
    for file in files:
        if file == "best.pt":
            print(os.path.join(root, file))
```
